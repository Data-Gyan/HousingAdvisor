```python
# remove 'property_id' column as it has no meaningful use for us
df.drop('property_id',axis=1, inplace=True)
```

```python
# a column name seems off the realty. Fix it
df.rename(columns = {'area': 'price_per_sqft'}, inplace=True)
df.head()
```

```python
import regex as re
df['society'].value_counts().shape
```

```python
# remove accidental inclusion of the society ratings from the society name
# May be a better strategy will be to keep this society overall rating as a new seperate column
df['society'] = df['society'].apply(lambda txt:re.sub(r'\d+(\.\d+)?\s*★.*$', '', str(txt)).strip()).str.lower()
```

```python
# will be meaningful to check if society names are not just misspelled or 
# written a bit differently for same society
# will check later though. Not doing it right now.
df['society'].value_counts().sort_index()

# df[df['society'] == 'ansal height 86'] = 'ansal heights 86'
```

```python
df = df[df['price'].str.lower() != 'price on request']
```

```python
df['price'].value_counts()
```

```python
def unify_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        elif x[1] == 'Lakh':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(unify_price)
```

```python
df['price_per_sqft'].value_counts()

# ₹ 10,000/sq.ft.
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.split().str.get(1).str.replace(',', '').str.strip().astype(float)
```

```python
df.info()
```

```python
# this column needs more extensive feature engineering.
# Leaving this as it is for now.
df['areaWithType']
```

```python
# check if there are null values in 'bedRoom' column
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype(np.int8)
```

```python
# df['bathroom'].isnull().sum()
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(np.int8)
```

```python
# Check for balconies
# Can not change it to integer as '3+' can be any number greater than 3
# Hence I am keeping it as an object. Will use it as categorical variable
df['balcony'].isnull().sum()
df['balcony'].value_counts()
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No', '0')
```

```python
df['additionalRoom'].isna().sum()
df['additionalRoom'].fillna('not available', inplace=True)
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['floorNum'].value_counts()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split().str.get(0).replace('Basement', '-1').replace('Ground', '0').replace('Lower', '0').str.extract(r'(\d+)')
```

```python
# df['facing'].value_counts()
df['facing'].fillna('NA', inplace=True)
```

```python
df['facing'].value_counts()
```

```python
df.insert(loc=4, column='area', value=round(df['price']*10000000/df['price_per_sqft']))
```

```python
df.insert(loc=1, column='property_type', value='flat')
```
