```python
# remove accidental inclusion of the society ratings from the society name
# May be a better strategy will be to keep this society overall rating as a new seperate column
df['society'] = df['society'].apply(lambda txt:re.sub(r'\d+(\.\d+)?\s*★.*$', '', str(txt)).strip()).str.lower()
```

```python
df = df[df['price'].str.lower() != 'price on request']
```

```python
df['price'].value_counts()
```

```python
def unify_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        elif x[1] == 'Lakh':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(unify_price)
```

```python
df['price_per_sqft'].value_counts()

# ₹ 10,000/sq.ft.
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.split().str.get(1).str.replace(',', '').str.strip().astype(float)
```
